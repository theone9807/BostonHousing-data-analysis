```python
import numpy as np
from numpy.random import randn
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import pandas as pd
```

```python
#reading the data from the storage using pandas
housing = pd.read_csv("housing.csv") # header =None
housing
```

```python
#plt.hist() is used to plot the histogram of the data
plt.hist(housing)
```

```python
#kernsl density estimation
xmin = housing["RM"].min()
xmax = housing["RM"].max()

x_axis = np.linspace(xmin, xmax, 100)
#bandwidht for estimation
bandwidth = ((4 * (housing["RM"].std()) ** 5) / (3 * len(housing["RM"]))) ** 0.2

kernal_list = []

for data_points in housing["RM"]:
    kernal = stats.norm(data_points, bandwidth).pdf(x_axis)
    kernal_list.append(kernal)
    
    kernal =kernal / kernal.max()
    kernal = kernal + 0.4
    
plt.plot(x_axis, kernal, color = "red", alpha = 0.5)
plt.ylim(0, 2)
```
